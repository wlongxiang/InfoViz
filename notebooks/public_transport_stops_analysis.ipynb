# Intro
This notebook processes the bus stops across netherlands

## Download data
  Run `get_public_transport_data.py` to download data in JSON format to `halts` directory

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
```

```python
from glob import glob
import json

all_stops = []
for json_file in sorted(glob("../halts/*.json")):
    j_dict = json.load(open(json_file))
    for one_stop in j_dict["features"]:
        all_stops.append(one_stop["attributes"])
```

```python
all_df = pd.DataFrame(all_stops)
all_df.sample(10)
```

```python
all_df.to_csv("../data/all_public_transport.csv", index=False)
```
