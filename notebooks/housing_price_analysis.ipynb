# Intro
In this notebook, we analyze the average housing price for 2018 in the Netherlands.
The data source comes from [CBS](https://www.cbs.nl/nl-nl/nieuws/2019/12/regionale-verschillen-in-huizenprijzen-groter#id=undefined)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
```

```python
city_house_prices_2019 = pd.read_csv("../data/gemiddelde-verkoopprijs-bestaande-koopwoningen-2019.csv")
city_house_prices_2019
```

```python
city_of_provinces = pd.read_excel("../data/Gemeenten-alfabetisch-2019.xls")
city_of_provinces
```

```python
results = pd.merge(city_house_prices_2019, city_of_provinces, left_on='statcode', right_on="GemeentecodeGM")
results
```

```python
results_final = results.drop(columns=["statcode", "GemeentecodeGM", "ProvinciecodePV"])
```

```python
print("top 10 most expensive areas 2019")
results_final.sort_values(["Gemiddelde verkoopprijs"],ascending=False)[:10]
```

```python
print("top 10 least expensive areas 2019")
results_final.sort_values(["Gemiddelde verkoopprijs"],ascending=True)[:10]
```

```python
results_final.to_csv("processed_housing_price_2019.csv", index=False)
```
