```python
import geopandas as gpd
```

```python
nl = gpd.read_file('Netherlands.shp')
```

```python
nl = nl.rename(columns={'Privincie': 'Provincie'})
```

```python
nl.to_file('Netherlands.shp')
nl.plot(column='Provincie')
```
